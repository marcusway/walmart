## Things to consider 

### Relationships between products might matter (I think). 
- It looks like the test data consists of whole missing days (for all products). The way that relationships between products could be leveraged would be somthing like the following: We want to know how many jars of peanut butter sold in LA on June 10. Peanut butter and jelly sales tend to be correlated. More jars of jelly than usual were sold at the Walmart in Santa Fe on June 10. Therefore, maybe more jars of peanut butter than usual were sold at the LA store. This probably won't be tremendously useful, though, since we also will have information on how many jars of peanut butter sold at the store in Santa Fe (and everywhere else), which will probably be the more useful metric. 

### Relationships between stores probably does matter. 
- Since the test data points will be from days around storms are particular stores, we can use data from similar stores on the same day as a sort of baseline prediction. This may be a very useful way of quantifying the impact of storms on the number of units sold for a particular item. If item A performs very similarly at Store 1 and Store 2, and there's a storm near Store 2 but not near Store 1, then the difference between the two stores during this period will be more informative than just the difference from baseline at Store 2 (I think)

### Some items are very seasonal. 
- There are some clearly seasonal items, which are only sold during certain months of the year. For these products, it would probably be safe to assign some "active period," and be sure to predict zero for the number of units sold for those items on any date outside the active period. 

### All of the test dates are around storms. 
- This could be very important. We don't really have to predict item performance on regular days at all (The different rows are different stores, the x-axis is time, dots are training data, orange dots are test data. All the orange dots are the six-day windows around weather events). However, the storms are surrounded by regular days, which will probably be used to give baseline. 

### Products aren't carried by all stores. 
- Almost all of the products are carried in fewer than 5 (of 45) stores. That might be a problem if we're trying to find clusters of similar stores, since the number of common items sold by any two stores is likely to be fairly small.
